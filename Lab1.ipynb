```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
#Anna Meleshko 319346045

# !wget https://www.fcbarcelonanoticias.com/uploads/s1/11/88/56/4/messi-250-goles-camp-nou_4_641x361.jpeg -O input.jpg
# image = plt.imread("./input.jpg")
image = cv2.imread('messi.jpg',cv2.IMREAD_COLOR) 

plt.figure(figsize = [12,5])
plt.subplot(131) #first img out of 3
plt.imshow(image)
plt.title("bgr")
plt.imshow(np.flip(image,axis = -1))

plt.subplot(132) 
plt.imshow(image)
plt.title("rgb")

plt.subplot(133)
plt.imshow(image)
plt.title("bgr")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2GRAY),cmap = 'gray')
plt.title("gray")
```

```python
img =np.zeros([600,600,3],dtype = np.uint8)
#drawing a line
img = cv2.line(img,(0,0),(599,599),(255,0,0),6)
#darwing circle
img = cv2.circle(img,(510,90),90,(0,0,255),-1)
#drawing rectangle
img = cv2.rectangle(img,(420,3),(598,180),(0,255,0),5)
#drawing elipse
img = cv2.ellipse(img,(320,320),(100,50),-50,0,180,(255,0,0),-1)
#drawing poligon
pts = np.array([[2,2],[250,60],[360,180],[200,280]],np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(img,[pts],True,(0,255,255),6)
#adding text
font = cv2.FONT_HERSHEY_COMPLEX
img = cv2.putText(img,'openCV',(10,580), font,4.8,(255,255,255),11,cv2.LINE_AA)
plt.imshow(img.astype(np.uint8))
```
